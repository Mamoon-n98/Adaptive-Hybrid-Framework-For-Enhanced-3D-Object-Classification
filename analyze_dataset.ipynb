```python
# Process starts: Analyze dataset.
import os
import glob
dataset_root = '/root/autodl-tmp/Adaptive-Hybrid-Framework-For-Enhanced-3D-Object-Classification/modelnet40/ModelNet40'
splits = ['train', 'test']
for split in splits:
    total_files = 0
    class_folders = sorted(os.listdir(dataset_root))
    for cls_name in class_folders:
        split_dir = os.path.join(dataset_root, cls_name, split)
        if not os.path.isdir(split_dir):
            print(f"Split {split} not found for class {cls_name}")
            continue
        off_files = glob.glob(os.path.join(split_dir, '*.off'))
        total_files += len(off_files)
        print(f"Class {cls_name} in {split}: {len(off_files)} .off files")
    print(f"Total .off files in {split}: {total_files}")
# Process ends: Analysis complete.
```
